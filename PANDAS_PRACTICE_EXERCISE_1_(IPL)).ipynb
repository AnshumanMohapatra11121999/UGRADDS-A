```python
# ipl17 contains the data for 2017
ipl17 = pd.DataFrame({'Team': ['MI', 'RPS', 'SRH', 'KKR', 'KXIP', 'DD', 'GL', 'RCB'],
                         'Matches': [14, 14, 14, 14, 14, 14, 14, 14],
                         'Won': [10, 9, 8, 8, 7, 6, 4, 3],
                         'Lost': [4, 5, 5, 6, 7, 8, 10, 10],
                         'Tied': [0, 0, 0, 0, 0, 0, 0, 0],
                         'N/R': [0, 0, 1, 0, 0, 0, 0, 1],
                         'NRR': [0.784, 0.176, 0.469, 0.641, 0.123, -0.512, -0.412, -1.299],
                         'For': [2407, 2180, 2221, 2329, 2207, 2219, 2406, 1845],
                         'Against': [2242, 2165, 2118, 2300, 2229, 2255, 2472, 2033]},
                         index = range(1,9)
                    )
# print the entire dataframe to check the entries
ipl17
```

```python

# Type your code here
#### Q3: Operations on multiple dataframes
'''
If all the stats are taken for both ‘ipl17’ and ‘ipl18’, which team with its total points greater than 25 will have the highest win percentage?

-  KKR
-  CSK
-  RPS
-  SRH
'''
# Type your code here
# Merge the two DataFrames based on 'Team' column
combined_df = pd.merge(ipl17, ipl18, on='Team', suffixes=('_17', '_18'))

```

```python
combined_df
```

```python
# Sum up the points from both DataFrames to get total points
combined_df['Total_Points'] = (combined_df['Won_17'] * 2) + combined_df['Tied_17'] + combined_df['N/R_17'] + (combined_df['Won_18'] * 2) + combined_df['Tied_18'] + combined_df['N/R_18']

combined_df
```

```python
# Filter out teams with total points greater than 25
teams_greater_than_25_points = combined_df[combined_df['Total_Points'] > 25]
teams_greater_than_25_points
```

```python
# Calculate win percentage for each team
teams_greater_than_25_points['Win_Percentage'] = (teams_greater_than_25_points['Won_17'] + teams_greater_than_25_points['Won_18']) / \
                                                 (teams_greater_than_25_points['Matches_17'] + teams_greater_than_25_points['Matches_18']) * 100


teams_greater_than_25_points
```

```python
# Find the team with the highest win percentage
team_highest_win_percentage = teams_greater_than_25_points.sort_values(by='Win_Percentage', ascending=False).iloc[0]['Team']

print("Team with the highest win percentage among those with total points greater than 25:", team_highest_win_percentage)
```
