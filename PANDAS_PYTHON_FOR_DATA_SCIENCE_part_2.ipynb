```python
'''
A pivot table is a handy tool to represent a dataframe in a structured and simplified manner.
It acts as an alternative to the groupby() function in Pandas. Pivot tables provide excel-like functionalities to create aggregate tables.

'''
```

```python
# Create a new dataframe
users= pd.DataFrame({"weekday" : ['Sun', 'Sun', 'Mon', 'Mon'],
                    "city": ['Austin', 'Dallas', 'Austin', 'Dallas'],
                    "visitors":[139, 237, 326, 456],
                    "signups" : [7, 12, 3, 5]})
```

```python
users
```

```python
users.set_index("weekday",inplace = True)
```

```python
users
```

```python
import matplotlib.pyplot as plt
users.plot(kind = "bar")
plt.show()
```

```python
visitors_pivot=users.pivot(columns = "city",values = "visitors")
visitors_pivot
```

```python
#df.pivot(columns='grouping_variable_col', values='value_to_aggregate', index='grouping_variable_row')
```

```python
visitors_pivot.plot(kind = "bar")
plt.show()
```

```python
visitors_signup=users.pivot(columns = "city")
visitors_signup
```

```python
visitors_signup.plot(kind="bar")
plt.show()
```

```python
#Subsetting and creating 2 different graphs for signup and visitors
visitors_signup["visitors"].plot(kind="bar")
visitors_signup["signups"].plot(kind="bar")
plt.show()

```

```python
'''When you have 2 sub sheets in a excel sheet
 for example in the salesreturn excel sheet you have 2 sub sheets called "Orders" and return
and inorder to select them indivisually we use sheet name attribute in read_excel function'''
```

```python
orders=pd.read_excel("/content/sales_returns.xlsx",sheet_name="Orders")
returns=pd.read_excel("/content/sales_returns.xlsx",sheet_name="Returns")
```

```python
orders.shape
```

```python
returns.shape
```

```python
orders.head()
```

```python
returns.head()
```

```python
#Lets merge the data frame based on a common factor
return_orders = orders.merge(returns,on = "Order ID")
return_orders.head()
#Here we are trying to merge the two data frame based on the factor order id
```

```python
return_orders.shape
#As we have done inner join so the common rows are present in the result dataset
```

```python
'''You can use the following command to merge two dataframes:

dataframe_1.merge(dataframe_2, on = ['column_1', 'column_2'], how = '____')'''
```

```python
#We want to see all the orers and their return status as yes or no in one dataframe
return_orders=orders.merge(returns,on = "Order ID",how = "left")
return_orders
```

```python
'''
The how attribute in the code above specifies the type of merge to be performed:

left: This will select the entries only in the first dataframe.
right: This will consider the entries only in the second dataframe.
outer: This takes the union of all the entries in the dataframes.
inner: This will result in the intersection of the keys from both frames.

SYNTX:MERGERD_DATAFRAME=FIRST DATAFRAME.MERGE(SECOND DATAFRAME,on = "cOMMON COLUMN",how="l/r/i/o")

'''
```

```python
# Dataframes having the same columns

df1 = pd.DataFrame({'Name': ['Aman', 'Joy', 'Rashmi', 'Saif'],
                    'Age' : ['34', '31', '22', '33'],
                    'Gender': ['M', 'M', 'F', 'M']}
                  )

df2 = pd.DataFrame({'Name': ['Akhil', 'Asha', 'Preeti'],
                    'Age' : ['31', '22', '23'],
                    'Gender': ['M', 'F', 'F']}
                  )
```

```python
df1.append(df2)
```

```python
# Another dataframe with same number of rows as df1
df3 = pd.DataFrame({'School': ['RK Public', 'JSP', 'Carmel Convent', 'St. Paul'],
                    'Graduation Marks': ['84', '89', '76', '91']}
                  )
```

```python
pd.concat([df1,df3])
```

```python
import pandas as pd
#Concatenate row basis (axis=0)
pd.concat([df1,df3],axis=0)
```

```python
#Concatenate on the basis of column (axis = 1)
pd.concat([df1,df3],axis=1)
```
